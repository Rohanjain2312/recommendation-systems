```python
import pandas as pd
import glob
import numpy as np
```

```python
files = glob.glob("data/*.csv")
```

```python
# fixing Amazon-Products.csv has index as first col
df = pd.read_csv("data/Amazon-Products.csv", index_col=0)
df.to_csv("data/Amazon-Products.csv", index=False)

# fixing bad values in rating and no_of_ratings
for file in files:
    df = pd.read_csv(file)
    df["ratings"] = pd.to_numeric(df["ratings"], downcast="float", errors='coerce')
    df["no_of_ratings"] = pd.to_numeric(df["no_of_ratings"], downcast="float", errors='coerce')
    df.to_csv(file, index=False)
```

```python
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df["discount_price"] = df["discount_price"].apply(lambda x: float(str(x).strip("₹").replace(",","")))
df["actual_price"] = df["actual_price"].apply(lambda x: float(str(x).strip("₹").replace(",","")))
df["ratings"].astype(float)
df["no_of_ratings"].astype(float)
df
```
